```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
#Dados normais
```

```python
#Criando uma variável com dados em uma distribuição normal com a função rvs (100 elementos)
dados = norm.rvs(size = 10000)
```

```python
#Histograma
plt.hist(dados, bins=20)
plt.title("Dados")
```

```python
#Gerando gráfico para verificar se a distribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
#Executando teste de shapiro
#segundo argumento é o valor p, não há como rejeitar a hipotese nula
stats.shapiro(dados)
```

```python
#Dados não normais
```

```python
dados2 = skewnorm.rvs(4,size=1000)
```

```python
#Histogramas
plt.hist(dados2, bins=20)
plt.title("Dados")
```

```python
#Geração de gráfico para verificar se a distribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(dados2)
```
